Outlier detection using z-score

```python
def get_zscore(value,mean,std):
    return (value-mean)/std
```

```python
df['zscore']=df.rev_mln.apply(lambda x: get_zscore(x,mean,std))
```

Outlier detection using mod z-score

```python
def get_mad(s):
    med=np.median(s)
    diff=abs(s-med)
    MAD=np.median(diff)
    return MAD
```

```python
MAD=get_mad(df.rev_mln)
```

```python
median=np.median(df['rev_mln'])
```

```python
MAD, median
```

General guideline for modified Z score is to use 3.5 as a thresold. i.e. anything that has a mod z score of 3.5 or more is an outlier

```python
def get_mod_zscore(x,MAD,median):
    return 0.6745*(x-median)/MAD
```

```python
df['mod_zscore']=get_mod_zscore(df.rev_mln,MAD,median)
```

```python
df.head()
```

```python
df_new2=df[(df.mod_zscore>-3.5)&(df.mod_zscore<3.5)]
```
