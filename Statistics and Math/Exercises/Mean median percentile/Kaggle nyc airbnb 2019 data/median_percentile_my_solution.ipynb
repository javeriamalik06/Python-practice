```python
df.info()
```

```python
df.price.describe()
```

```python
q1=df.price.quantile(0.25)
q3=df.price.quantile(0.75)
```

```python
iqr=q3-q1
print(iqr)
```

```python
outlier_ul=df.price.mean()+(1.5*iqr)
outlier_ll=df.price.mean()-(1.5*iqr)
```

```python
df[df.price>=outlier_ul]
```

```python
df_new=df[(df.price<=outlier_ul)&(df.price>=outlier_ll)]
```

```python
df_new.shape
```

```python
df_new.price.describe()
```

```python
df_new.sample(5)
```
