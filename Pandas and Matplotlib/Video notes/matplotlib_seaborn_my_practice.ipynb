```python
import pandas as pd
df_sales=pd.read_excel('linechart.xlsx')
df_sales.head()
```

```python
#importing library
from matplotlib import pyplot as plt
```

```python
#chart is big so to optimize it before plotting(x,y)-x is width and y is height
plt.figure(figsize=(12,4))
#making a line chart (x-axis,y-axis)
plt.plot(df_sales.Quarter,df_sales.Fridge)
#to remove the trash line above chart
plt.show()
```

```python
#Charting fridge, dishwasher and washing machine revenue together
plt.figure(figsize=(12,4))
#specifying the color and label of line this time
plt.plot(df_sales.Quarter,df_sales.Fridge, color='orange',label='Fridge')
plt.plot(df_sales['Quarter'],df_sales['Dishwasher'],color='green',label='Dishwasher')
plt.plot(df_sales['Quarter'],df_sales['Washing Machine'],color='black',label='Washing Machine')

#Adding title and labels to the plot
plt.title('Product Sales')
plt.ylabel('Revenue (mln $)')
plt.xlabel('Financial Quarter')
#to show the labels
plt.legend()

plt.show()
```

```python
total_sales=df_sales[['Fridge','Dishwasher','Washing Machine']].sum()
total_sales
```

```python
#Making a pie chart and showing percentage and labels
plt.pie(total_sales, labels=total_sales.index, autopct='%1.2f%%')
plt.show()
```

```python
#also making the pies distant and adding shadow, start angle rotates the chart
plt.pie(total_sales, labels=total_sales.index, autopct='%1.2f%%', explode=(0.1,0,0), shadow=True, startangle=140)
plt.show()
```

```python
#Plotting a bar chart, directly on the dataframe
#Index was the numbers hence we have to set x=Quarter, otherwise we wont include it
df_sales.plot(kind='bar', x='Quarter')

#to rotate the xlabel
plt.xticks(rotation=50)

plt.ylabel('Revenue (mln $)')
plt.xlabel('Financial Quarter')
```
